# Diamond Price Prediction with Deep Learning

Bu projede, Elmas fiyatlarını tahmin etmek için PyTorch kullanarak bir Derin Öğrenme (Deep Learning) modeli geliştireceğiz.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import warnings
warnings.filterwarnings("ignore")

data = pd.read_csv("data/diamonds.csv")
print(data.head())
```

## Veri Temizleme

```python
# Boyutsuz (hatalı) verileri temizleme
data = data.drop(data[data["x"]==0].index)
data = data.drop(data[data["y"]==0].index)
data = data.drop(data[data["z"]==0].index)

# Aykırı değerleri filtreleme
data = data[(data["depth"]<75)&(data["depth"]>45)]
data = data[(data["table"]<80)&(data["table"]>40)]
data = data[(data["x"]<30)]
data = data[(data["y"]<30)]
data = data[(data["z"]<30)&(data["z"]>2)]

print("Temizlenmiş veri boyutu:", data.shape)
```

## Veri Ön İşleme (One-Hot Encoding ve Scaling)
Derin öğrenme modelleri için kategorik verileri One-Hot Encoding ile dönüştürüyor ve tüm verileri ölçeklendiriyoruz.

```python
# Hedef ve Özelliklerin Ayrılması
x = data.drop('price', axis=1)
y = data[['price']]

# One-Hot Encoding
x = pd.get_dummies(x, drop_first=True)
print("One-Hot Encoding sonrası sütun sayısı:", x.shape[1])

# Eğitim ve Test Bölme
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Standardizasyon (Scaling)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# PyTorch Tensor Dönüşümü
x_train_tensor = torch.tensor(x_train, dtype=torch.float32)
x_test_tensor = torch.tensor(x_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32)
```

## Yapay Sinir Ağı (Neural Network) Modeli
Çok katmanlı bir sinir ağı oluşturuyoruz.

```python
# Model Tanımı
class NeuralNetwork(nn.Module):
    def __init__(self, input_dim):
        super(NeuralNetwork, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 128)
        self.fc3 = nn.Linear(128, 256)
        self.fc4 = nn.Linear(256, 128)
        self.fc5 = nn.Linear(128, 64)
        self.fc6 = nn.Linear(64, 32)
        self.fc7 = nn.Linear(32, 1)

    def forward(self, x):
        x = nn.ReLU()(self.fc1(x))
        x = nn.ReLU()(self.fc2(x))
        x = nn.ReLU()(self.fc3(x))
        x = nn.ReLU()(self.fc4(x))
        x = nn.ReLU()(self.fc5(x))
        x = nn.ReLU()(self.fc6(x))
        return self.fc7(x)

# Model, Loss ve Optimizer Kurulumu
input_dim = x_train.shape[1]
model = NeuralNetwork(input_dim)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

## Model Eğitimi (Training Loop)
Modeli eğitiyor ve Early Stopping ile aşırı öğrenmeyi (overfitting) engelliyoruz.

```python
num_epochs = 150
batch_size = 32

# DataLoader oluşturma
train_data = torch.utils.data.TensorDataset(x_train_tensor, y_train_tensor)
train_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size, shuffle=True)

# Early Stopping değişkenleri
best_loss = float('inf')
patience = 10
patience_counter = 0
history = {'loss': [], 'val_loss': []}

print("Eğitim Başlıyor...")
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    for batch_x, batch_y in train_loader:
        optimizer.zero_grad()
        outputs = model(batch_x)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    
    avg_epoch_loss = epoch_loss / len(train_loader)
    
    # Validation
    model.eval()
    with torch.no_grad():
        val_outputs = model(x_test_tensor)
        val_loss = criterion(val_outputs, y_test_tensor).item()
    
    history['loss'].append(avg_epoch_loss)
    history['val_loss'].append(val_loss)

    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {avg_epoch_loss:.4f}, Val Loss: {val_loss:.4f}')
    
    # Early Stopping Kontrolü
    if val_loss < best_loss:
        best_loss = val_loss
        patience_counter = 0
        # En iyi modeli kaydetmek istersek burada wights kaydedilebilir
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"Early stopping triggered at Epoch {epoch+1}.")
            break
```

## Eğitim Grafiği (Loss Curve)

```python
plt.figure(figsize=(10, 5))
plt.plot(history['loss'], label='Training Loss')
plt.plot(history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.show()
```

## Model Değerlendirmesi

```python
model.eval()
with torch.no_grad():
    predictions = model(x_test_tensor).numpy()
    actuals = y_test_tensor.numpy()

print("R^2 Score:", metrics.r2_score(actuals, predictions))
print("MAE:", metrics.mean_absolute_error(actuals, predictions))
print("MSE:", metrics.mean_squared_error(actuals, predictions))
print("RMSE:", np.sqrt(metrics.mean_squared_error(actuals, predictions)))
```

## Sonuç ve İyileştirmeler

Bu notebook'ta yapılan geliştirmeler:

1.  **PyTorch Entegrasyonu**: Klasik Makine Öğrenmesi (sklearn) modelleri yerine, özelleştirilebilir bir Derin Öğrenme modeli (Neural Network) entegre edildi.
2.  **One-Hot Encoding**: Kategorik değişkenler (`cut`, `color`, `clarity`) için daha uygun olan One-Hot Encoding yöntemi kullanıldı (önceki versiyonda Label Encoding kullanılıyordu). Bu, modelin kategorik ilişkileri daha doğru öğrenmesini sağlar.
3.  **Modern Mimarisi**: 6 Gizli Katmanlı (Hidden Layers), ReLU aktivasyon fonksiyonlu bir yapay sinir ağı tasarlandı.
4.  **Early Stopping**: Modelin aşırı öğrenmesini (overfitting) engellemek ve en iyi yerde eğitimi durdurmak için Early Stopping mekanizması eklendi.
5.  **Detaylı Değerlendirme**: Loss grafikleri ve metrikler (R2, RMSE) ile model başarısı detaylandırıldı.
