# Mobile Price Classification

Bu projede, cep telefonlarının teknik özelliklerine göre fiyat aralıklarını (0-3 arası) tahmin eden makine öğrenimi modelleri geliştirip karşılaştırıyoruz.

Bu notebook, **algo_test** fonksiyonu kullanarak çoklu classification model karşılaştırması yapmaktadır.

Amaç: Telefonların RAM, pil kapasitesi, ekran boyutu gibi özelliklerine bakarak hangi fiyat kategorisine gireceğini tahmin etmek ve en iyi modeli bulmaktır.

## 1. Kütüphanelerin Yüklenmesi

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

%matplotlib inline
```

## 2. algo_test Fonksiyonu Tanımı

```python
gnb = GaussianNB()
l = LogisticRegression(max_iter=1000)
d = DecisionTreeClassifier()
r = RandomForestClassifier()
gb = GradientBoostingClassifier()
kn = KNeighborsClassifier()
ab = AdaBoostClassifier()
svc = SVC()

def algo_test(x, y):
    modeller = [gnb, l, d, r, gb, kn, ab, svc]
    isimler = ["GaussianNB", "LogisticRegression", "DecisionTreeClassifier", 
               "RandomForestClassifier", "GradientBoostingClassifier", "KNeighborsClassifier",
               "AdaBoostClassifier", "SVC"]

    x_train, x_test, y_train, y_test = train_test_split(x, y.ravel(), test_size=.20, random_state=42)
    
    # Feature Scaling (özellikle SVC ve KNN için önemli)
    scaler = StandardScaler()
    x_train_scaled = scaler.fit_transform(x_train)
    x_test_scaled = scaler.transform(x_test)
    
    accuracy = []
    precision = []
    recall = []
    f1 = []
    mdl = []

    print("Veriler hazır modeller deneniyor")
    for model in modeller:
        print(model, " modeli eğitiliyor!..")
        model = model.fit(x_train_scaled, y_train)
        tahmin = model.predict(x_test_scaled)
        mdl.append(model)
        accuracy.append(accuracy_score(y_test, tahmin))
        # Çok sınıflı classification için average="weighted" kullanıyoruz
        precision.append(precision_score(y_test, tahmin, average="weighted"))
        recall.append(recall_score(y_test, tahmin, average="weighted"))
        f1.append(f1_score(y_test, tahmin, average="weighted"))
        print(confusion_matrix(y_test, tahmin))

    print("Eğitim tamamlandı.")
    
    metrics = pd.DataFrame(columns=["Accuracy", "Precision", "Recall", "F1", "Model"], index=isimler)
    metrics["Accuracy"] = accuracy
    metrics["Precision"] = precision  
    metrics["Recall"] = recall
    metrics["F1"] = f1
    metrics["Model"] = mdl

    metrics.sort_values("F1", ascending=False, inplace=True)

    print("En başarılı model: ", metrics.iloc[0].name)
    model = metrics.iloc[0, -1]
    
    # En iyi model için test
    scaler = StandardScaler()
    x_train_scaled = scaler.fit_transform(x_train)
    x_test_scaled = scaler.transform(x_test)
    tahmin = model.predict(x_test_scaled)
    
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, tahmin))
    print("Classification Report:")
    print(classification_report(y_test, tahmin))
    print("Diğer Modeller:")
    
    return metrics.drop("Model", axis=1)
```

## 3. Veri Yükleme

```python
data = pd.read_csv("data/mobile_prices.csv")
print(f"Veri boyutu: {data.shape}")
data.head()
```

## 5. Özellik ve Hedef Değişken Tanımlama

```python
# Özellikler (X) - price_range hariç tüm sütunlar
X = np.array(data.drop("price_range", axis=1))

# Hedef değişken (y)
y = np.array(data[["price_range"]])

print(f"✓ X boyutu: {X.shape}")
print(f"✓ y boyutu: {y.shape}")
print(f"✓ Özellik sayısı: {X.shape[1]}")
```

## 6. algo_test ile Tüm Modelleri Çalıştır

```python
results = algo_test(X, y)
results
```

## 7. Sonuç Görselleştirme

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Accuracy
ax1 = axes[0, 0]
results["Accuracy"].plot(kind="barh", ax=ax1, color="steelblue")
ax1.set_title("Model Accuracy Karşılaştırması", fontsize=12)
ax1.set_xlabel("Accuracy")

# Precision
ax2 = axes[0, 1]
results["Precision"].plot(kind="barh", ax=ax2, color="green")
ax2.set_title("Model Precision Karşılaştırması", fontsize=12)
ax2.set_xlabel("Precision")

# Recall
ax3 = axes[1, 0]
results["Recall"].plot(kind="barh", ax=ax3, color="orange")
ax3.set_title("Model Recall Karşılaştırması", fontsize=12)
ax3.set_xlabel("Recall")

# F1
ax4 = axes[1, 1]
results["F1"].plot(kind="barh", ax=ax4, color="crimson")
ax4.set_title("Model F1-Score Karşılaştırması", fontsize=12)
ax4.set_xlabel("F1-Score")

plt.tight_layout()
plt.suptitle("Mobile Price Classification - Model Performans Karşılaştırması", fontsize=14, y=1.02)
plt.show()
```

3. **DecisionTreeClassifier** - Karar ağacı
4. **RandomForestClassifier** - Rastgele orman
5. **GradientBoostingClassifier** - Gradient boosting

7. **AdaBoostClassifier** - AdaBoost

Performans ölçütleri olarak **Accuracy**, **Precision**, **Recall** ve **F1-Score** kullanıldı. Bu çok sınıflı (4 sınıf: 0, 1, 2, 3) bir classification problemi olduğundan, precision, recall ve F1-score için `weighted average` kullanıldı.
