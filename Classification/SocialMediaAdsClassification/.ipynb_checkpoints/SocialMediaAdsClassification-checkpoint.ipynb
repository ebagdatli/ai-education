# Social Media Ads Classification

Bu işlem, ürününüzü satın alma ihtimali en yüksek ve en karlı müşteri kitlesini belirlemek için reklam verilerinin analiz edilmesidir. Her ürün her yaş ve gelir grubuna hitap etmez; örneğin, gençlerin telefon kılıfı harcamaları orta yaşlılara göre daha fazladır. Özetle; reklam bütçesini doğru kişilere yönlendirerek verimi artırmayı hedefler.

Bu notebook, sosyal medya reklamları veri seti üzerinde **algo_test** fonksiyonu kullanarak çoklu classification model karşılaştırması yapmaktadır.

Amaç: Kullanıcıların yaşlarına ve tahmini gelirlerine bakarak bir ürünü satın alıp almayacaklarını tahmin eden makine öğrenimi modelleri geliştirmek ve karşılaştırmaktır.

## 1. Kütüphanelerin Yüklenmesi

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

%matplotlib inline
```

## 2. algo_test Fonksiyonu Tanımı

```python
gnb = GaussianNB()
l = LogisticRegression()
d = DecisionTreeClassifier()
r = RandomForestClassifier()
gb = GradientBoostingClassifier()
kn = KNeighborsClassifier()
ab = AdaBoostClassifier()
svc = SVC()

def algo_test(x, y):
    modeller = [gnb, l, d, r, gb, kn, ab, svc]
    isimler = ["GaussianNB", "LogisticRegression", "DecisionTreeClassifier", 
               "RandomForestClassifier", "GradientBoostingClassifier", "KNeighborsClassifier",
               "AdaBoostClassifier", "SVC"]

    x_train, x_test, y_train, y_test = train_test_split(x, y.ravel(), test_size=.20, random_state=42)
    
    # Feature Scaling (özellikle SVC ve KNN için önemli)
    scaler = StandardScaler()
    x_train_scaled = scaler.fit_transform(x_train)
    x_test_scaled = scaler.transform(x_test)
    
    accuracy = []
    precision = []
    recall = []
    f1 = []
    mdl = []

    print("Veriler hazır modeller deneniyor")
    for model in modeller:
        print(model, " modeli eğitiliyor!..")
        model = model.fit(x_train_scaled, y_train)
        tahmin = model.predict(x_test_scaled)
        mdl.append(model)
        accuracy.append(accuracy_score(y_test, tahmin))
        precision.append(precision_score(y_test, tahmin, average="binary"))
        recall.append(recall_score(y_test, tahmin, average="binary"))
        f1.append(f1_score(y_test, tahmin, average="binary"))
        print(confusion_matrix(y_test, tahmin))

    print("Eğitim tamamlandı.")
    
    metrics = pd.DataFrame(columns=["Accuracy", "Precision", "Recall", "F1", "Model"], index=isimler)
    metrics["Accuracy"] = accuracy
    metrics["Precision"] = precision  
    metrics["Recall"] = recall
    metrics["F1"] = f1
    metrics["Model"] = mdl

    metrics.sort_values("F1", ascending=False, inplace=True)

    print("En başarılı model: ", metrics.iloc[0].name)
    model = metrics.iloc[0, -1]
    
    # En iyi model için test
    scaler = StandardScaler()
    x_train_scaled = scaler.fit_transform(x_train)
    x_test_scaled = scaler.transform(x_test)
    tahmin = model.predict(x_test_scaled)
    
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, tahmin))
    print("Classification Report:")
    print(classification_report(y_test, tahmin))
    print("Diğer Modeller:")
    
    return metrics.drop("Model", axis=1)
```

## 3. Veri Yükleme

```python
data = pd.read_csv("data/social.csv")
print(f"Veri boyutu: {data.shape}")
data.head()
```

```python
print("Veri istatistikleri:")
data.describe()
```

```python
print("Hedef sınıflar:")
print(data["Purchased"].unique())
print("\nSınıf dağılımı:")
print(data["Purchased"].value_counts())
```

## 4. Veri Görselleştirme

```python
plt.figure(figsize=(15, 10))
plt.title("Product Purchased By People Through Social Media Marketing")
sns.histplot(x="Age", hue="Purchased", data=data)
plt.show()
```

```python
plt.title("Product Purchased By People According to Their Income")
sns.histplot(x="EstimatedSalary", hue="Purchased", data=data)
plt.show()
```

## 5. Özellik ve Hedef Değişken Tanımlama

```python
# Özellikler (X)
X = np.array(data[["Age", "EstimatedSalary"]])

# Hedef değişken (y)
y = np.array(data[["Purchased"]])

print(f"✓ X boyutu: {X.shape}")
print(f"✓ y boyutu: {y.shape}")
print(f"✓ Özellikler: ['Age', 'EstimatedSalary']")
```

## 6. algo_test ile Tüm Modelleri Çalıştır

```python
results = algo_test(X, y)
results
```

## 7. Sonuç Görselleştirme

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Accuracy
ax1 = axes[0, 0]
ax1.barh(results.index, results['Accuracy'], color='steelblue', alpha=0.7)
ax1.set_xlabel('Accuracy')
ax1.set_title('Model Accuracy Karşılaştırması')
ax1.set_xlim(0, 1)

# Precision
ax2 = axes[0, 1]
ax2.barh(results.index, results['Precision'], color='green', alpha=0.7)
ax2.set_xlabel('Precision')
ax2.set_title('Model Precision Karşılaştırması')
ax2.set_xlim(0, 1)

# Recall
ax3 = axes[1, 0]
ax3.barh(results.index, results['Recall'], color='orange', alpha=0.7)
ax3.set_xlabel('Recall')
ax3.set_title('Model Recall Karşılaştırması')
ax3.set_xlim(0, 1)

# F1 Score
ax4 = axes[1, 1]
ax4.barh(results.index, results['F1'], color='red', alpha=0.7)
ax4.set_xlabel('F1 Score')
ax4.set_title('Model F1 Score Karşılaştırması')
ax4.set_xlim(0, 1)

plt.tight_layout()
plt.show()
```

## Açıklama: Yapılan İşlemler

Bu notebook'ta aşağıdaki adımlar gerçekleştirilmiştir:

### 1. Veri Yükleme
- **social.csv** dosyası yüklendi
- 400 örnek, 3 özellik içeriyor (Age, EstimatedSalary, Purchased)

### 2. Özellikler (Features)
- **Age**: Kullanıcının yaşı
- **EstimatedSalary**: Tahmini maaş

### 3. Hedef Değişken (Target)
- **Purchased**: Ürün satın alındı mı?
  - 0: Satın almadı
  - 1: Satın aldı

### 4. algo_test Fonksiyonu
Aşağıdaki 8 classification modeli tek bir fonksiyon çağrısıyla test edildi:

| Model | Açıklama |
|-------|----------|
| **GaussianNB** | Gaussian Naive Bayes - Sürekli özellikler için |
| **LogisticRegression** | Lojistik Regresyon - Lineer sınıflandırıcı |
| **DecisionTreeClassifier** | Karar Ağacı - Kural tabanlı |
| **RandomForestClassifier** | Random Forest - Topluluk yöntemi |
| **GradientBoostingClassifier** | Gradient Boosting - Topluluk yöntemi |
| **KNeighborsClassifier** | K-En Yakın Komşu |
| **AdaBoostClassifier** | AdaBoost - Adaptive Boosting |
| **SVC** | Support Vector Classifier |

### 5. Değerlendirme Metrikleri

| Metrik | Açıklama |
|--------|----------|
| **Accuracy** | Doğru tahmin oranı (TP+TN)/(Total) |
| **Precision** | Pozitif tahminlerin doğruluğu TP/(TP+FP) |
| **Recall** | Gerçek pozitiflerin yakalanma oranı TP/(TP+FN) |
| **F1 Score** | Precision ve Recall'un harmonik ortalaması |

### 6. Sonuçlar
- Tüm modeller F1 Score değerine göre sıralanarak karşılaştırmalı tablo ve görselleştirme ile sunulmuştur
- En başarılı model için detaylı Confusion Matrix ve Classification Report gösterilmiştir
- Binary classification problemi olduğundan average="binary" parametresi kullanılmıştır
- Feature Scaling (StandardScaler) uygulanarak özellikle SVC ve KNN modellerinin performansı artırılmıştır
