# Music Genres Clustering

Bu projede, Spotify veri setini kullanarak müzik türlerini kümeleyeceğiz. 
Yapılacak iyileştirmeler:
1. Özelliklerin doğru şekilde ölçeklendirilmesi (Standard Scaler).
2. Elbow Yöntemi ile optimal küme sayısının belirlenmesi.
3. K-Means kümeleme ve 3D Görselleştirme.
4. Hiyerarşik Kümeleme (Dendrogram) analizi.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as shc
import plotly.graph_objects as go
import warnings
warnings.filterwarnings("ignore")

# Veriyi Yükle
data = pd.read_csv("data/Spotify-2000.csv")
data = data.drop("Index", axis=1)
print(data.head())
```

## Veri Analizi ve Korelasyon

```python
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
plt.figure(figsize=(12, 10))
sns.heatmap(data[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## Özellik Seçimi ve Ölçeklendirme
Kümeleme için sayısal özellikleri seçiyoruz ve `StandardScaler` ile ölçeklendiriyoruz.

```python
# İlgili sayısal sütunları seçelim
features = ["Beats Per Minute (BPM)", "Loudness (dB)", "Liveness", 
            "Valence", "Acousticness", "Speechiness", "Energy", "Danceability"]

data_numeric = data[features]

# Ölçeklendirme (StandardScaler)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_numeric)
```

## Elbow Yöntemi (Optimal Küme Sayısı)
Hangi K değerinin en iyi sonucu verdiğini görmek için Inertia değerlerini çiziyoruz.

```python
wcss = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 15), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Inertia)')
plt.show()
```

## K-Means Kümeleme
Elbow grafiğine veya kullanım amacımıza göre 10 küme oluşturuyoruz (Orijinal projedeki gibi).

```python
kmeans = KMeans(n_clusters=10, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(data_scaled)
data["Music Segments"] = clusters

# Kümeleri isimlendirelim (Opsiyonel)
data["Cluster Name"] = data["Music Segments"].apply(lambda x: f"Cluster {x+1}")
```

## 3D Görselleştirme
Kümeleri BPM, Energy ve Danceability eksenlerinde görselleştiriyoruz.

```python
PLOT = go.Figure()
for i in list(data["Music Segments"].unique()):
    PLOT.add_trace(go.Scatter3d(
        x = data[data["Music Segments"]== i]['Beats Per Minute (BPM)'],
        y = data[data["Music Segments"] == i]['Energy'],
        z = data[data["Music Segments"] == i]['Danceability'],                        
        mode = 'markers',
        marker_size = 5,
        marker_line_width = 1,
        name = f"Cluster {i+1}"
    ))

PLOT.update_traces(hovertemplate='BPM: %{x} <br>Energy: %{y} <br>Danceability: %{z}')
PLOT.update_layout(
    width = 800, 
    height = 800, 
    autosize = True, 
    showlegend = True,
    scene = dict(
        xaxis=dict(title = 'Beats Per Minute (BPM)'),
        yaxis=dict(title = 'Energy'),
        zaxis=dict(title = 'Danceability')
    ),
    title = "Music Genres Clustering 3D"
)
PLOT.show()
```

## Dendrogram (Hiyerarşik Kümeleme)
Veri setinin hiyerarşik yapısını görmek için Dendrogram çizimi.

```python
plt.figure(figsize=(15, 8))
plt.title("Music Genres Dendrogram")
# Ward metodu varyansı minimize eder
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.xlabel('Songs')
plt.ylabel('Euclidean Distances')
plt.show()
```

## Sonuç ve İyileştirmeler

Bu projede yapılan iyileştirmeler:
1.  **Ölçeklendirme Düzeltildi**: `StandardScaler` kullanarak verileri ölçeklendirdik. Bu, K-Means gibi mesafe temelli algoritmalar için kritiktir. Önceki kodda ölçeklendirme uygulanmıyordu.
2.  **Özellik Seçimi**: Kümeleme için daha fazla ve anlamlı özellik (`Energy`, `Danceability` vb.) dahil edildi.
3.  **Elbow Yöntemi**: Optimal küme sayısını belirlemek için görsel bir araç eklendi.
4.  **Dendrogram**: Verilerin hiyerarşik ilişkisini gösteren Dendrogram grafiği eklendi.
5.  **Kod Temizliği**: Gereksiz döngüler kaldırıldı ve kod daha okunabilir hale getirildi.
